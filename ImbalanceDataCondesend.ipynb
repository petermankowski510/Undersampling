```python
"""
Condensed Nearest Neighbor Rule
Author: Peter Mankowski"""
from numpy import where
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.under_sampling import CondensedNearestNeighbour
from matplotlib import pyplot
```

```python
# Simple, fake dataset
n=10000
nf=2
w=0.99
X, y = make_classification(n_samples=n, n_features=nf, n_redundant=0,n_clusters_per_class=1, 
                           weights=[w], flip_y=0, random_state=1)
# Counter to verify quantity of datapoints(original)
counter = Counter(y)
print("Original, imbalanced set:", counter)
```

```python
"""Undersampling def."""
undersample = CondensedNearestNeighbour(n_neighbors=1)
X, y = undersample.fit_resample(X, y)
# Counter to verify quantity of datapoints(original)
counter = Counter(y)
print("Undersampled set:", counter)
```

```python
"""plot"""
for label, _ in counter.items():
    row_ix = where(y == label)[0]
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```
